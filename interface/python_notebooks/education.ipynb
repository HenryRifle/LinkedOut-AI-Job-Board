```python
data_2019=pd.read_excel("D:\\Rutgersstuff\\BigDataAlgorithms\\BigDataProjectData\\2019-29\\education.xlsx", sheet_name=[1,2,3], skiprows = lambda x: x == 0)
```

```python
data_2019[1].head(10)
```

```python
data_2019[1].columns = ["entry_level_degree","emp","emp_dist","emp_change","median_wage_2020"]
```

```python
data_2019[1][1:-4].hvplot.bar(x="entry_level_degree",y="median_wage_2020", color="skyblue", height=400, yformatter=BasicTickFormatter(use_scientific=False)).opts(xrotation=90, bgcolor="yellow", title="Median Wage Distribution")
```

```python
emp_dist_plot=data_2019[1][1:-4].set_index("entry_level_degree").plot.pie(y="emp_dist", figsize=(8,8))
emp_dist_plot.set_ylabel("")
emp_dist_plot.set_title("Employment Distribution")
emp_dist_plot.legend(bbox_to_anchor=(1.05,1), fontsize=9).set_title("Entry Level Degrees")
emp_dist_plot
```
