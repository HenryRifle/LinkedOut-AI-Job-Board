```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load the data from users.xlsx
file_path = "interface/project_data/generated_data/users.xlsx"
df = pd.read_excel(file_path)

# Drop non-skill columns ('Name' and 'Added Skills')
skill_columns = df.drop(columns=['Name', 'Added Skills'], errors='ignore')
skill_columns

```

```python
skill_columns= skill_columns.fillna(0)
skill_columns
```

```python
#cosine similarity implemented
cos_sim_matrix = cosine_similarity(skill_columns)
```

```python
# Create a DataFrame for the similarity matrix, including user names
similarity_df_filled = pd.DataFrame(cos_sim_matrix, index=df['Name'], columns=df['Name'])
similarity_df_filled
```
