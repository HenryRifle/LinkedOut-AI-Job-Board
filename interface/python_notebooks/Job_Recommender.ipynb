```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import random
```

```python
skills_data.head()
```

```python
#Occupation Recommendations

user_skills = []

for i in range(0,18):
    user_skills.append(random.uniform(1,5))
    

cosine_similarities = cosine_similarity([user_skills], skills_data.astype(float))


skills_df['Similarity'] = cosine_similarities[0]
recommended_occupations = skills_df[['Occupation', 'Similarity']].sort_values(by='Similarity', ascending=False)
recommended_occupations.head()
```

```python
#Created artificial jobs dataset matching the format of skills dataset for easier computation.
jobs_data = pd.read_csv('generated_data\\jobs.csv')
```

```python
jobs_data.head()
```

```python
# Create new df without Job Title and Company Name columns
jobs_skills = jobs_data.drop(['Job Title','Company Name'],axis=1) 
```

```python
#Actual Job Listing Recommendations

user_skills = []

#random user
for i in range(0,18):
    user_skills.append(random.uniform(1,5))

cosine_similarities = cosine_similarity([user_skills], jobs_skills)

jobs_data['Similarity'] = cosine_similarities[0]
recommended_occupations = jobs_data[['Job Title','Company Name','Similarity']].sort_values(by='Similarity', ascending=False)
recommended_occupations.head()
```
