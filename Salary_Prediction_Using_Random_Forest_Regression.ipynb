```python
# skills_df = skills_df[skills_df['Salary'] == '>=239,200']
# skills_df = skills_df[skills_df['Salary'] == '—']


# skills_df['Salary'] = skills_df['Salary'].str.replace(r'\D', '', regex=True)
skills_df['Salary'] = skills_df['Salary'].replace('—', '0')
skills_df['Salary'] = skills_df['Salary'].replace('>=239,200', '239200')
skills_df['Salary'] = skills_df['Salary'].astype(int)

# skills_df['Salary'] = skills_df['Salary'].replace('', pd.NA)
# skills_df['Salary'] = skills_df['Salary'].fillna(0).astype(int)
```

```python
X = skills_df.drop(['Salary','Occupation'],axis = 1)
y = skills_df['Salary']
```

```python
y.unique()
```

```python
skills_df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state= 1)
```

```python
y_test
```

```python
model = RandomForestRegressor(n_estimators = 50, random_state = 20)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
rmse = mse ** 0.5
rmse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
scores =[]
for k in range(1, 200):
    rfc = RandomForestRegressor(n_estimators=k)
    rfc.fit(X_train, y_train)
    y_pred = rfc.predict(X_test)
    scores.append(r2_score(y_test, y_pred))

import matplotlib.pyplot as plt
%matplotlib inline

# plot the relationship between K and testing accuracy
# plt.plot(x_axis, y_axis)
plt.plot(range(1, 200), scores)
plt.xlabel('Value of n_estimators for Random Forest Classifier')
plt.ylabel('Testing Accuracy')

```
